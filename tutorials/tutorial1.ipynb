# Technical Analysis of Cryptocurrencies

Basic tutorial.

```python
import crypto_ta as cta

import warnings
warnings.filterwarnings("ignore") # just for the sake of tutorial, in general not recommended
```

## Setting up

Create connection to API with data first.

**Remember to set up your API_KEY and API_SECRET first!**

```python
api = cta.BinanceAPI()
# api = cta.BinanceAPI(api_key='xxx', api_secret='xxx') # alternative way
```

Create a `CryptoTA` object that governs all the computations.

```python
ta = cta.CryptoTA(api)
```

We need to fetch the data for a specific cyrpto pair from the server first. Let's analyse BTC to USDT for 2021 year weekly.

```python
ta.get_ohlcv('BTCUSDT', interval='1w', from_date='2021-01-01')
```

## Basic plots

We can display the Open High Low Close and Volume values...

```python
ta.ohlcv.head()
```

and plot them with moving averages.

```python
ta.plot()
```

Candle plot.

```python
ta.plot_candle()
```

We can also plot general trend line.

```python
ta.plot_trend_lines()
```

## Technical Analysis

Now, let's compute some less trivial Technical Analysis measures.

```python
ta.calculate_ta()
```

We can take a sneak peak at calculated features.

```python
ta.features.head()
```

TA measures are added in columns, to see a full list of performed operations we can use:

```python
ta.get_ta_features()
```

And we can easily vizualize any of them. Here let's take a look at the Volume Force Index (FI).

```python
ta.plot_ta('volume_fi')
```

## Predictions

CryptoTA offers simple predictions method based on ARIMA model.
- https://en.wikipedia.org/wiki/Autoregressive_integrated_moving_average

Here we compute ARMA model with AR order 4 and MA order 2.

```python
ta.calculate_arima(order = (4,0,2), show_diag=True)
```

Now we can try to predict BTC to USD price for the coming 5 weeks.

```python
from datetime import datetime, timedelta

delta = timedelta(days=5*7)
```

```python
end = ta.ohlcv.index[-1] + delta
ta.plot_forecasts(end)
```

Like you can see, the error is large, so we shouldn't really trust these predictions (but our ARIMA parameters pick was quite arbitrary). Anyway, you can still forecast future values with our model for the *n* time units with the following command:

```python
ta.forecast(5)
```
